```python
import pandas as pd

df = pd.read_excel('faltas_treinamento_sujo.xlsx')

display(df)
```

Exercício 1 – Higienização Básica


```python
# a) Verifique valores nulos e tipos incorretos

coluna_string = df.select_dtypes(include=['string','object']).columns.to_list()
print(coluna_string)

coluna_int = df.select_dtypes(include=['int']).columns.to_list()
print(coluna_int)

print(df.info())

display(df)
```

```python
# b) Corrija a coluna de datas inválidas.

df['Data_Ultima_Falta'] = pd.to_datetime(df['Data_Ultima_Falta'], format='%d/%m/%Y', errors='coerce')

display(df)
```

```python
# c) Normalize os nomes dos turnos e dias da semana.

df['Turno'] = df['Turno'].str.strip().str.lower().str.title().replace({'Manha': 'Manhã'})
df['Dia_da_Semana_Ultima_Falta'] = df['Dia_da_Semana_Ultima_Falta'].str.strip().str.lower().str.title()

display(df)
```

Exercício 2 – Remoção e Substituição

```python
# a) Remova os registros duplicados.

df = df.drop_duplicates()

print(df.duplicated().sum())
```

```python
# b) Substitua os valores nulos da coluna Evadiu por “Desconhecido”.


df['Evadiu'] = df['Evadiu'].fillna('Desconhecido')

display(df[80:90])
```

```python
# c) Substitua os valores nulos da coluna Data_Ultima_Falta pela data mais comum.

df['Data_Ultima_Falta'] = df['Data_Ultima_Falta'].fillna(df['Data_Ultima_Falta'].mode()[0])
display(df[80:90])
```
